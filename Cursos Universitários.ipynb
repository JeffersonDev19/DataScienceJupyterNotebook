```python
from matplotlib import pyplot as plt

%matplotlib notebook
lst = []
sd = []
index = []
count = 1
lst = df.query('mensalidade > 900 and uf_busca == "SP" and nota_integral_ampla > 700')['nome']
for x in (lst):
    if x not in sd:
        sd.append(x)
        index.append(count)
        count += 1

lsNum = []
for x in (sd):
    count = 0
    for y in (lst):
        if x == y:
            count = count+1
    lsNum.append(count)

for x in range(len(sd)):
    print(str(index[x]) + " - " + sd[x])
    
plt.bar(index,lsNum)
plt.xlabel('Números Identificador dos Cursos')
plt.ylabel('Quantidade existente de cada Curso')
plt.show()
print("Aqui foi feito a pesquisa dos melhores cursos em relação nota total do curso, em cima dos maiores valores")
```
